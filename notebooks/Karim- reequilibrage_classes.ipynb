```python

import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from imblearn.over_sampling import SMOTE

X_train = pd.read_csv("../data/raw/X_train.csv", sep=",",index_col=0)
y_train = pd.read_csv("../data/raw/Y_train.csv", sep=",",index_col=0)

texts = X_train["designation"]
labels = y_train

def text_smote(texts, labels, random_state=42):
    vectorizer = TfidfVectorizer(ngram_range=(1,2))
    X = vectorizer.fit_transform(texts)
    
    smote = SMOTE(random_state=random_state)
    X_res, y_res = smote.fit_resample(X, labels)
    
    return X_res, y_res, vectorizer

# Appliquer text_smote
X_resampled, y_resampled, vectorizer = text_smote(texts, labels)
```

```python
# Afficher les distributions avant et après
print("Distribution originale :")
print(labels.value_counts(normalize=False))
print("\nDistribution après SMOTE-NC :")
print(y_resampled.value_counts(normalize=False))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Diviser les données pour l'entraînement
X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled, 
    test_size=0.2, 
    random_state=42,
    stratify=y_resampled
)

# Entraîner un modèle
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)

# Évaluer le modèle
y_pred = clf.predict(X_test)
print("\nRapport de classification :")
print(classification_report(y_test, y_pred))
```
